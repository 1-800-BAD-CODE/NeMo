```python
import pandas as pd
from nemo.collections.common.tokenizers.column_coder import ColumnCodes
from omegaconf import OmegaConf
import pickle
```

```python
df = pd.read_csv('./inflation.csv',parse_dates=['Date'])
columns = list(df.columns)
df['Month'] = df['Date'].dt.month
columns[0] = 'Month'
df
```

```python
tab_structure = []
for c in columns:
    if c != 'Month':
        item = {
            "name": c,
            "code_type": "float",
            "args": {
                "code_len": 3,  # number of tokens used to code the column
                "base": 32,   # the positional base number. ie. it uses 32 tokens for one digit
                "fillall": True, # whether to use full base number for each token or derive it from the data.
                "hasnan": False, # can it handles nan or not
            }
        }
    else:
        item = {
            "name": c,
            "code_type": "category",
        }
    tab_structure.append(item)
print(OmegaConf.to_yaml(tab_structure))
print(columns)
```

```python
example_arrays = {}
for col in tab_structure:
    col_name = col['name']
    example_arrays[col_name] = df[col_name].dropna().unique()
cc = ColumnCodes.get_column_codes(tab_structure, example_arrays)
with open('table.pickle', 'wb') as handle:
    pickle.dump(cc, handle)
```

```python
seq_length = 1024
tokens_per_line = sum(cc.sizes) + 1

window_size = seq_length // tokens_per_line
print(f"tokens per line {tokens_per_line}, windows size {window_size}")
```

```python
model_df = df[columns]
model_df
```

```python
import json
import functools
from multiprocessing import Pool

delimiter = ','
eod_str = '<|endoftext|>'


def get_docs_from_df(inputs):
    df, skip = inputs
    num_cols = len(df.columns)
    skip_rows = skip // num_cols
    skip_column = skip % num_cols
    # print(skip_rows, skip_column)
    full_msgs = []
    count = 0
    for row in df.iterrows():
        count += 1
        if count - 1 < skip_rows:
            continue
        items = row[1].values
        str_items = [str(int(items[i])) if i==0 else str(items[i]) for i in range(skip_column, len(items))]
        full_msgs.append(delimiter.join(str_items)) 
        skip_column = 0
    # use end of line to seperate rows
    text = '\n'.join(full_msgs)
    text += eod_str
    text_doc = {'text': text}
    doc = json.dumps(text_doc)+'\n'
    return [doc]


def gen_one_doc(model_df, n_cores, length):
    pool = Pool(n_cores)
    docs = pool.map(get_docs_from_df, [(model_df, i) for i in range(length)])
    pool.close()
    pool.join()
    return functools.reduce(lambda a, b: a + b, docs)


def generate_docs(df, filename, length):
    with open(filename, 'w') as f:
        docs = gen_one_doc(df, 10, length)
        for doc in docs:
            f.write(doc)
```

```python
train_index = df['Date'] < '2020-01-01'
val_index = (df['Date'] >= '2020-01-01') & (df['Date'] < '2021-01-01')
test_index = df['Date'] >= '2021-01-01'
assert train_index.sum() + val_index.sum() + test_index.sum() == len(df)
```

```python
train_df = model_df[train_index]
val_df = model_df[val_index]
test_df = model_df[test_index]
print(len(train_df), len(val_df), len(test_df))

generate_docs(train_df, 'train.jl', seq_length)
generate_docs(val_df, 'val.jl', 1)
generate_docs(test_df, 'test.jl', 1)



len(model_df.columns)
```
